```python
from sklearn.preprocessing import LabelEncoder
for column in Employee.columns:
    if Employee[column].dtype==np.number:
        continue
    else:
        Employee[column]=LabelEncoder().fit_transform(Employee[column])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=1,criterion='entropy',random_state=0)
```

```python
x=Employee.drop(['Yes'],axis=1)
y=Employee['Yes']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=0)
x_train.head()
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```

```python
pr=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pr)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest=RandomForestClassifier(n_estimators=1, criterion = 'entropy',random_state=0)
```

```python
forest.fit(x_train,y_train)
```

```python
print(x_train)
```

```python
score=forest.score(x_train,y_train)
print('randomforest classifier',np.abs(score)*100)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,forest.predict(x_test))
TP=cm[0][0]
TN=cm[1][1]
FP=cm[1][0]
FN=cm[0][1]
print(cm)
print('Model Testing Accuracy - ',(TP+TN)/(TP+TN+FP+FN))
```
